```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import glob
import os
import feather
sns.set(style="whitegrid")
```

--------------------------------------

### Preparing the frames to store information

```python
datasets = ['Adiac', 'Fish', 'OliveOil', 'Phoneme', 'ShapesAll', 'SwedishLeaf', 'WordSynonyms']

# row corresponds to classifier, col to dataset
errors = pd.DataFrame(columns=datasets)

regressors = ['RR_sklearn', 
              'RR_catch22',
              'RR_kats',
              'RR_tsfeatures',
              'RR_tsfel',
              'RR_tsfresh']
```

```python
catch22    = pd.read_feather('featureFrames/output_catch22.feather')
kats       = pd.read_feather('featureFrames/output_kats.feather')
tsfeatures = pd.read_feather('featureFrames/output_tsfeatures.feather')
tsfel      = pd.read_feather('featureFrames/output_tsfel.feather')
tsfresh    = pd.read_feather('featureFrames/output_tsfresh.feather')

featF = [catch22, kats, tsfeatures, tsfel, tsfresh]

rawTimeSeries = []

for dataset in datasets:
    rawTimeSeries.append(pd.read_feather('data/' + dataset + '_plain.feather')) #add data column as well?
```

--------------------------------------

### SKLEARN

```python
from sklearn import linear_model
clf = linear_model.BayesianRidge()
```

```python
err = []

for i,dataset in zip(range(len(datasets)), datasets):
    
    print(dataset)
    
    X = rawTimeSeries[i].iloc[:-1,:-1].copy()
    y = rawTimeSeries[i].iloc[:-1,-1].copy()
    X_toPredict = rawTimeSeries[i].iloc[-1,:-1].copy()
    y_toPredict = rawTimeSeries[i].iloc[-1,-1].copy()
    
    clf.fit(X, y)
    
    pred = clf.predict([X_toPredict])
    
    err.append(abs(y_toPredict-pred[0]))

errors.loc[0] = err
```

```python
errors
```

--------------------------------------

### sklearn x features

```python
j = 1
for frame in featF:
    
    err = []
    print('Round ', j+1)
    
    for i,dataset in zip(range(len(datasets)), datasets):
        
        print(dataset)
        
        df = frame[frame.data == dataset].copy()
    
        X = df.iloc[:-1,:-2].copy()
        y = df.iloc[:-1,-1].copy()
        X_toPredict = df.iloc[-1,:-2].copy()
        y_toPredict = df.iloc[-1,-1].copy()
    
        clf.fit(X, y)
    
        pred = clf.predict([X_toPredict])
    
        err.append(abs(y_toPredict-pred[0]))

    errors.loc[j] = err
    j += 1
    print('\n')
    
```

### Plotting error

```python
def plotPerDataset(axes, y_pos, nbrPlots, df):
    
    for c in range(nbrPlots):
        
        axes[c].set_title(df.columns[c])
        #axes[c].set_xlabel('Regressor')
        #axes[c].set_ylabel('Error')
        
        # Create bars
        color = ['black'] + (len(df)-1)*['maroon']
        axes[c].bar(y_pos, df.iloc[:, c], color=color)
        
        axes[c].set_xticks(y_pos, regressors, rotation='vertical')
        
        
    
    
def plotPerClassifier(axes, y_pos, nbrPlots, df):
    
    for c in range(nbrPlots):
        
        axes[c].set_title(regressors[c])
        #axes[c].set_xlabel('Dataset')
        #axes[c].set_ylabel('Error')
        
        color = 'maroon'
        
        if c == 0:
            color = 'black'
        
        # Create bars
        axes[c].bar(y_pos, df.iloc[c], color=color)
        
        #
        axes[c].set_xticks(y_pos, datasets, rotation='vertical') 
        
        
        
    
```

```python
def plotError(df, perDataset = True):
    
    cols = 3
    
    if perDataset:
        nbrPlots = len(df.columns)
        y_pos    = np.arange(len(df))
    else:
        nbrPlots = len(df)
        y_pos    = np.arange(len(df.columns))
    
    rows = int(np.ceil(nbrPlots/cols))
    
    fig, ax = plt.subplots(rows, cols, figsize=(30, 30))
    # using padding
    fig.tight_layout(pad=14.0)
    
    
    axes = ax.flatten()
    fs = 20
    
    if perDataset:
        plotPerDataset(axes, y_pos, nbrPlots, df)
        fig.text(0.5, 0.04, 'Regressor', ha='center', fontsize=fs)
        fig.text(0.04, 0.5, 'Absolute Error', va='center', rotation='vertical', fontsize=fs)
    else:
        plotPerClassifier(axes, y_pos, nbrPlots, df)
        fig.text(0.5, 0.04, 'Dataset', ha='center', fontsize=fs)
        fig.text(0.04, 0.5, 'Absolute Error', va='center', rotation='vertical', fontsize=fs)
    
        
    
    
     
```

#### Plot for each dataset

```python
plotError(errors)
```

#### Plot for each classifier

```python
plotError(errors, False)
```
